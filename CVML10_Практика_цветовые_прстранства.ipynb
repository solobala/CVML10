# Цветовые пространства

## Загрузка изображения из файла

```python
import cv2
```

```python
img = cv2.imread('/content/lena.png', cv2.IMREAD_COLOR) # cv2.IMREAD_COLOR - Значение по умолчанию . =1
```

```python
print('type: ', type(img))
print('shape: ', img.shape)
```

## Визуализация в matplotlib

```python
%matplotlib inline
from matplotlib import pyplot as plt
plt.imshow(img)
```

```python
# изменяем порядок каналов для визуализации (b, g, r) -> (r, g, b)
# cv2.cvtColor - ф-я для конвертации. 2-й аргумент - во что хотим конвертировать
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

## Переходим в пространство цветов HSV

```python
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
# Каналы Hue и Saturation - ногсители цвета. value - насыщенность
# в hue разные цвета имеют разные оттенки
plt.figure(1, figsize = (16, 4))
plt.subplot(141)
plt.imshow(img[...,::-1])
plt.subplot(142)
plt.imshow(img_hsv[...,0], cmap='gray')
plt.title('Hue')
plt.subplot(143)
plt.imshow(img_hsv[...,1], cmap='gray')
plt.title('Saturation')
plt.subplot(144)
plt.imshow(img_hsv[...,2], cmap='gray')
plt.title('Value');
```

```python
plt.imshow(img_hsv) # При визуализации в формае hsv matplotlib ее не знает эту схему и трактует как красный зеленый синий
```
